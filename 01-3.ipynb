```python
smelt_df = df[df['Species'] == 'Smelt']
smelt_df.head()
```

```python
smelt_length = smelt_df['Length2'].tolist()
smelt_weight = smelt_df['Weight'].tolist()
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
```

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
fish_data = [[l, w]for l, w in zip(length, weight)]
fish_data
```

```python
target_data = [1] * len(bream_length) + [0] * len(smelt_length)
```

```python
from sklearn.neighbors import KNeighborsClassifier #클래스
```

```python
kn = KNeighborsClassifier() #인스턴스화
```

```python
kn.fit(fish_data, target_data)
```

```python
kn.score(fish_data, target_data)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(25, 400, c='red', s=100, marker='*')
plt.scatter(16, 30, s=100, marker='h')
plt.scatter(smelt_length, smelt_weight)
plt.show()
```

```python
kn49 = KNeighborsClassifier(49)
kn49.fit(fish_data, target_data)
```

```python
kn49.score(fish_data, target_data) #= 35/49 (bream만 맞춤)
```

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, target_data)

#가 Kn~함수 기본값
for n in range(5,50):
    kn.n_neighbors = n
    score = kn.score(fish_data, target_data)
  
    print(n, score)
    if score < 1:
        break
```

## iris 데이터 연습

```python
import seaborn as sns

df = sns.load_dataset('iris')
df
```

1. 데이터 가져오기
2. 어떤 특성을 사용할지
3. KN 학습
4. matplotlib으로 시각화
5. 예측

```python
df['species'].unique()
```

```python
df.shape
```

```python
setosa_df = df[df['species'] == 'setosa']
versi_df = df[df['species'] == 'versicolor']
vergi_df = df[df['species'] == 'virginica']

setosa_df['petal_length'].nunique()
vergi_df['petal_length'].nunique()
```

```python
setosa_plength = setosa_df['petal_length'].tolist()
setosa_pwidth = setosa_df['petal_width'].tolist()

versi_plength = versi_df['petal_length'].tolist()
versi_pwidth = versi_df['petal_width'].tolist()

vergi_plength = vergi_df['petal_length'].tolist()
vergi_pwidth = vergi_df['petal_width'].tolist()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(setosa_plength, setosa_pwidth, c='red')
plt.scatter(versi_plength, versi_pwidth, c='blue')
plt.scatter(vergi_plength, vergi_pwidth, c='green')
plt.xlabel('petal length')
plt.ylabel('petal width')

plt.scatter(4.0, 1.0, c='yellow', marker='*')
```

```python
length = setosa_plength + vergi_plength
width = setosa_pwidth + vergi_pwidth
```

```python
flower_data = [[l, w]for l, w in zip (length, width)]
```

```python
target_data = [1] * len(setosa_plength) + [0] * len(vergi_plength)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(29)
```

```python
kn.fit(flower_data, target_data)
```

```python
kn.predict([[4, 1]])
```

```python
for n in range(5, 30):
    kn.n_neighbors = n
    score = kn.score(flower_data, target_data)

    print(n, score)
    if score < 1:
        break
```
